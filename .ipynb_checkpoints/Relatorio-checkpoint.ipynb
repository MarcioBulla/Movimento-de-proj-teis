# Questão 1
As condições iniciais são: $x(0)=z(0)=0$ e $\vec{v}(0)=v_0\hat{v}$.

Sendo $\vec{v_0} = v_0\cos{\theta}\hat{\imath} + v_0\sin{\theta}\hat{k}$

## Alternativa a

Para resolver esta questão precisamos desmembrar a o vetor posição em suas componetes.
$$ m \devd{\vec{r}}{t} =-mg\hat{k} - b\vec{v_0}$$

Desta forma temos duas EDO de segunda ordem.

$$ \devd{x}{t} = - \frac{b}{m} \dev{x}{t}$$
$$ \devd{z}{t} =-g - \frac{b}{m} \dev{z}{t}$$

### Resolvendo a EDO em $x$
Usaremos as função do scipe *solve_ivp*.

#### Definindo função da EDO em $x$ para resolver:

```python
def x_t(t, x, b, m):
    return [x[1], - b/m * x[1]]
```

Definindo condições iniciais e constantes:

```python
## condiçÕes iniciais ##
x_0 = 0
v_0 = 30
theta = np.pi/4
vx_0 = v_0*np.cos(theta)
## constantes ##
b = 3
m = 5
```

Resolvendo a EDO numericamente com o metodo LSODA:

```python
sol_x = solve_ivp(fun=x_t, t_span=[0, 5], y0=[x_0, vx_0], args=(
    b, m), t_eval=np.linspace(0, 5, 100), method="LSODA")
```

Plotando grafico da solução para velecidade e posição:

```python
fig, ((x), (vx)) = plt.subplots(2, 1, sharex=True)
x.plot(sol_x.t, sol_x.y[0], label="Posição em fução do tempo", c="b")
x.set_ylabel("Posição (m)")
x.set_title("Posição em função do tempo")
x.legend()
vx.plot(sol_x.t, sol_x.y[1], label="Velocidade em função do tempo", c="orange")
vx.set_xlabel("Tempo (s)")
vx.set_ylabel(r"Velocidade ($\frac{m}{s}$)")
vx.set_title("Velocidade em função do tempo")
vx.legend()
```

Grafico da velocidade em função do tempo, note que esta favoravel com o que esperamos com uma velociade tendendo a zero ea posição estaginando quando isto ocorre.

### Resolvendo para $z$

Note que muita coisas vamos reaproveita como as condição unicial de $v_0$ e todas as constantes já determinadas.

#### Definindo função para $z$:

```python
def z_t(t, z, g, b, m):
    return [z[1], - g - b/m * z[1]]
```

Definindo a constante faltante e condiçoes iniciais:

```python
## Condiões iniciais ##
z_0 = 0
vz_0 = v_0*np.sin(theta)
## Constantes ##
g = 9.8
```

Resolvendo EDO da mesma forma que foi resolvida para a função $x$:

```python
sol_z = solve_ivp(fun=z_t, t_span=[0, 5], y0=[z_0, vz_0], args=(
    g, b, m), t_eval=np.linspace(0, 5, 100), method="LSODA")
```

Plotando grafico da posição verticalem função do tempo:

```python
fig, ((ax1),(ax2)) = plt.subplots(2,1, sharex = True)
ax1.plot(sol_z.t, sol_z.y[0], c="b")
ax1.set_ylabel("Posição (m)")
ax1.set_title("Prosição")

ax2.plot(sol_z.t, sol_z.y[1], c="orange")
ax2.set_ylabel(r"velocidade ($\frac{m}{s}$)")
ax2.set_title("Prosição")
ax2.set_xlabel("tempo (t)")
```

Perceba que chegamos no que esperamos novamente. Pois, diferente de $x$, $z$ parte de zero e chega novamente em $0$.

### Alternativa b

Para fazer varias soluçoes diferentes usaremos um *loop* onde trocaremos somente os angulos $\theta$ em um intervalo de $[20; 70]$ com 10 pontos. Desta forma, temos:

```python
for _ in np.linspace(20, 70, 10, endpoint=True):  # loop variando $\theta$
    # Tranformando graus em radianos
    theta_i = np.radians(_)
    # Condições iniciais ara cada angulação
    vx_0i = v_0*np.cos(theta_i)
    vz_0i = v_0*np.sin(theta_i)
    # Solução para cada eixo
    sol_xi = solve_ivp(fun=x_t, t_span=[0, 5], y0=[x_0, vx_0i], args=(
        b, m), t_eval=np.linspace(0, 5, 100), method="LSODA")
    sol_zi = solve_ivp(fun=z_t, t_span=[0, 5], y0=[z_0, vz_0i], args=(
        g, b, m), t_eval=np.linspace(0, 5, 100), method="LSODA")
    # Plotagem do grafico para cada theta
    plt.plot(sol_xi.y[0], sol_zi.y[0], label=r"$\theta _i$=" +
             f"{np.degrees(theta_i) :.0f}°")

plt.xlim(0, 35)
plt.ylim(0, 20)
plt.title("Posição para diferentes valores de θ")
plt.xlabel("Horizontal (m)")
plt.ylabel("Vertical (m)")
plt.legend()
```

### Alternativa c

Podemos agrupar estes pontos usando a opção *event* da função *solve_ivp*, no qual agrupa onde os pontos são zero. Desta forma, pegamos os ponto onde $\dev{z}{t}=0$, ou seja o poto onde $z$ é máximo; $z=0$ onde a particula esta no chão, logo podemos pegar o maximo alcance. 

Definindo as funções onde queremos capturar os pontos:

```python
def topo(t, z, g, b, m):
    return z[1]


def alcance(t, z, g, b, m):
    return z[0]
```

Fazendo soluções para varios $\theta$ igual o que foi feito no grafico anterio, porem com mais pontos:

```python
pontos = np.array([[], [], []])  # Criando array para recolher os pontos

for _ in np.linspace(0, 180, 200, endpoint=True):  # loop variando $\theta$
    # Tranformando graus em radianos
    theta_i = np.radians(_)
    # Condições iniciais para cada angulação
    vz_0i = v_0*np.sin(theta_i)
    vx_0i = v_0*np.cos(theta_i)
    # Solução do eixos x e z
    sol_zi = solve_ivp(fun=z_t, t_span=[0, 5], y0=[z_0, vz_0i], args=(
        g, b, m), t_eval=np.linspace(0, 5, 100), method="LSODA", events=(topo, alcance), dense_output=True)
    sol_xi = solve_ivp(fun=x_t, t_span=[0, 5], y0=[x_0, vx_0i], args=(
        b, m), t_eval=sol_zi.t_events[1], method="LSODA")  # note que pegamos somente os tempo das raizes
    # Recolhendo pontos do evento topo e alcance
    pontos = np.append(pontos, [[theta_i], sol_zi.y_events[0][:, 0], [
                       sol_xi.y[0][1] if _ not in [0,90, 180] else sol_xi.y[0][0]]], 1)
```

Plotando grafico da altura e alcance maximo em função do angulo:

```python
fig, ((ax1), (ax2)) = plt.subplots(2, 1, sharex=True)

ax1.plot(pontos[0], pontos[1], c="r")
ax1.set_xlim(pontos[0][0], pontos[0][-1])
ax1.set_ylim(0, 25)
ax1.set_ylabel("Altura máxima (m)")

ax2.plot(pontos[0], pontos[2], c="b")
ax2.set_xlim(pontos[0][0], pontos[0][-1])
ax2.set_ylim(-35, 35)
ax2.set_xlabel("Ângulo (radians)")
ax2.set_ylabel("Alcance máximo (m)")
```

# Questão 2
Semelhante a questão anterior porem agora com vento e a constante ao quadrado da resistencia do ar. Desta forma, temos:
$$ $$

As condições iniciais são: $x(0)=z(0)=0$ e $\vec{v}(0)=v_0\hat{v}$.

Sendo $\vec{v_0} = v_0\cos{\theta}\hat{\imath} + v_0\sin{\theta}\hat{k}$
## Alternativa a

Para resolver esta questão precisamos desmembrar a o vetor posição em suas componetes.
$$ m \devd{\vec{r}}{t} =-mg\hat{k} - b^2 (\vec{v_0}+\vec{v_w})$$
onde $\vec{v_w} = u\hat{\imath} + w\hat{k}$

Desta forma temos duas EDO de segunda ordem.

$$ \devd{x}{t} = - \frac{b^2}{m} \left(\dev{x}{t} + u\right)$$
$$ \devd{z}{t} =-g - \frac{b^2}{m} \left(\dev{z}{t} + w\right)$$


#### Definindo função da EDO em $x$ para resolver:

```python
def xu_t(t, x, b, m, u):
    return [x[1], - np.power(b,2)/m * (x[1]+u)]
```

Definindo condições iniciais e constantes:

```python
# CondiçÕes iniciais
x_0 = 0
v_0 = 30
theta = np.pi/4
vx_0 = v_0*np.cos(theta)
# Constantes
b = 3
m = 5
u = 1
```

Resolvendo a EDO numericamente com o metodo LSODA:

```python
sol_xu = solve_ivp(fun=xu_t, t_span=[0, 5], y0=[x_0, vx_0], args=(
    b, m, u), t_eval=np.linspace(0, 5, 100), method="LSODA")
```

Plotando grafico da solução para valecidade e posição:

```python
fig, ((x), (vx)) = plt.subplots(2, 1, sharex=True)
x.plot(sol_xu.t, sol_xu.y[0], label="Posição em fução do tempo", c="b")
x.set_ylabel("Posição (m)")
x.set_title("Posição em função do tempo")
x.legend()
vx.plot(sol_xu.t, sol_xu.y[1], label="Velocidade em função do tempo", c="orange")
vx.set_xlabel("Tempo (s)")
vx.set_ylabel("Velocidade (m)")
vx.set_title("Velocidade em função do tempo")
vx.legend()
```

Grafico da velocidade em função do tempo, note que esta favoravel com o que esperamos com uma velociade tendendo a uma velocidade terminal mais rapidamente, por causa do vento contrario e a posição retrocendendo.
